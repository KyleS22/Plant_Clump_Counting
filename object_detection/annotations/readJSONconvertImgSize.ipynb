```python
import os
import json
```

```python
JSONINFILENAME  = "032018_annotations.json"
JSONOUTFILENAME = "032018_annotations_new.json"

with open(JSONINFILENAME, "r") as read_file:
    JSONdata = json.load(read_file)
```

```python
CLASSES = ["original", "clumped", "isolated"]
```

```python
HEIGHT = 500
WIDTH  = 4000

if HEIGHT > WIDTH: # TODO: this is not implemented (yet?) because for "annotation" dataset width > height
    NUMPATCHES = HEIGHT // WIDTH
    NEWHEIGHT  = WIDTH
    NEWWIDTH   = WIDTH
    INDEXARR   = list(range(0,HEIGHT+1,WIDTH))
elif WIDTH > HEIGHT:
    NUMPATCHES = WIDTH // HEIGHT
    NEWHEIGHT  = HEIGHT
    NEWWIDTH   = HEIGHT
    INDEXARR   = list(range(0,WIDTH+1,HEIGHT))
else:
    NUMPATCHES = 1
    NEWHEIGHT  = HEIGHT
    NEWWIDTH   = WIDTH
    INDEXARR   = []
    print(f"WARNING: Height is the same as width ({NEWHEIGHT}).")

PATCHCOUNTARR = list(range(NUMPATCHES))

print(f"New img: {NUMPATCHES} patches at dims. {NEWWIDTH}x{NEWHEIGHT}")
print(INDEXARR)
print(PATCHCOUNTARR)
```

```python
IMG_FORMAT = ".tif"
```

```python
for filename, data in JSONdata.items():
    print(f"***** PROCESSING {filename} *****")
    for class_ in CLASSES:
        print(f"{class_}: {data[class_]}")
```

```python
newJSON = {}
```

```python
for filename, data in JSONdata.items():
    print(f"***** PROCESSING {filename} *****")
    basename = os.path.splitext(os.path.basename(filename))[0]
    for i in PATCHCOUNTARR:
        newKey = f"{basename}_{i}{IMG_FORMAT}"
        newJSON[newKey] = {}
        for class_ in CLASSES:
            newJSON[newKey][class_] = []
```

```python
for filename, data in newJSON.items():
    print(f"***** PROCESSING {filename} *****")
    for class_ in CLASSES:
        print(f"{class_}: {data[class_]}")
```

```python
# TODO: Configure this with more than just width > height mode?

for filename, data in JSONdata.items():
    print(f"***** PROCESSING {filename} *****")
    basename = os.path.splitext(os.path.basename(filename))[0]
    for class_ in CLASSES:
        for datum in data[class_]:
            T = datum[0]
            L = datum[1]
            B = datum[2]
            R = datum[3]
            
            for i in PATCHCOUNTARR:
                leftIdx  = INDEXARR[i]
                rightIdx = INDEXARR[i+1]
                shift    = NEWWIDTH*i
                # Case 1: "Normal" case: bbox is only in 1 patch
                if L >= leftIdx and R < rightIdx:
                    newJSON[f"{basename}_{i}{IMG_FORMAT}"][class_].append([T, L-shift, B, R-shift])
                # Case 2: "Bad" case: bbox is in two patches
                elif L >= leftIdx and L < rightIdx and R > rightIdx:
                    newJSON[f"{basename}_{i}{IMG_FORMAT}"][class_].append([T, L-shift, B, rightIdx-1-shift])
                    if i != PATCHCOUNTARR[-1]: # Ideally, this would never happen on the last patch anyway, buuuut...
                        newJSON[f"{basename}_{i+1}{IMG_FORMAT}"][class_].append([T, rightIdx-shift-NEWWIDTH, B, R-shift-NEWWIDTH]) # Additional NEWWIDTH subtract because the +1 patch
```

```python
for filename, data in newJSON.items():
    print(f"***** PROCESSING {filename} *****")
    for class_ in CLASSES:
        print(f"{class_}: {data[class_]}")
```

```python
with open(JSONOUTFILENAME, 'w') as f:
    json.dump(newJSON, f)
```
