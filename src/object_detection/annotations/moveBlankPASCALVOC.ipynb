```python
import os
import glob
import xml.etree.cElementTree as ET
```

```python
CWD = os.getcwd()

infolder = os.path.join(CWD, "pascalvoc_labels_patched_oneclass")
outfolder = os.path.join(infolder, "blanks")

if not os.path.exists(outfolder):
    os.makedirs(outfolder)
    print("WARNING: Making output dir, {}".format(outfolder))
```

```python
infiles = glob.glob(infolder + "/*.xml")

#print(infiles) # DEBUG
```

```python
emptyFiles = []

for file_ in infiles:
    print(f"***** PROCESSING {file_} *****")
    tempTree = ET.parse(file_)
    tempRoot = tempTree.getroot()
    if not tempRoot.findall('object'):
        emptyFiles.append(file_)
        print("File is empty.")

print("Empty:", emptyFiles)
```

```python
def moveFiles(fileList, newFolder):
    for filepath in fileList:
        smallFilename = os.path.basename(filepath)
        os.rename(filepath, os.path.join(newFolder, smallFilename))
```

```python
moveFiles(emptyFiles, outfolder)
```
