```python
import cv2
import os
import glob

import numpy as np

from math import ceil
```

```python
INFILEEXT  = ".png"
OUTFILEEXT = INFILEEXT#".png"

IMAGEFOLDER  = "D:\\tensorflow\\projects\\PlantClumpDetection\\annotations\\output_stitch\\"
EXPORTFOLDER = "D:\\tensorflow\\projects\\PlantClumpDetection\\annotations\\output_stitch_patched\\"
```

```python
inImg = glob.glob(IMAGEFOLDER+'*'+INFILEEXT)

print(inImg)
```

```python
def get_avg_colour(image):
    avg_color_per_row = np.average(image, axis=0)
    avg_color = np.average(avg_color_per_row, axis=0)
    return tuple(avg_color)
```

```python
for filename in inImg:
    print("***** PROCESSING {} *****".format(filename))
    frame = cv2.imread(filename)
    if frame is None:
        print("WARNING: {} cannot be read. Skipping...".format(filename))
        continue
    
    basename = os.path.splitext(os.path.basename(filename))[0]
    output_folder = os.path.join(EXPORTFOLDER, basename)
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
        print("WARNING: Making folder, {}".format(output_folder))
    
    h, w, c = frame.shape
    
    numPatches    = int(ceil(w/h)) # Ceil 'cause we don't want to discard last information
    indexArr      = list(range(0,numPatches*h+1,h))
    patchCountArr = list(range(numPatches))
    
    replicate = cv2.copyMakeBorder(frame,0,0,0,numPatches*h-w,cv2.BORDER_CONSTANT,value=get_avg_colour(frame[0:10,-10:-1,:])) # pad border to make perfect squares
    
    for i in patchCountArr:
        leftIdx  = indexArr[i]
        rightIdx = indexArr[i+1]
        outFilename = "{}_{}{}".format(basename, i, OUTFILEEXT)
        exportPath = os.path.join(output_folder, outFilename)
        print("Writing {}...".format(exportPath))
        cv2.imwrite(exportPath, replicate[:,leftIdx:rightIdx,:])
    #break # DEBUG
```
